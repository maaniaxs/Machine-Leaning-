```python

from sklearn.model_selection import train_test_split

from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# IN TRAINING PROCESS
# finding best random_state

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

max_acu= 0
max_rs= 0
for r_state in range(50,400):
    x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = r_state,test_size=0.20,stratify=y)
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    accu = accuracy_score(y_test,y_pred)
        #print("random state :",r_state," and f1 score: ",tmp)
        
    if accu>max_acu:
        max_acu=accu
        max_rs= r_state
        print("maximum accuracy :",max_acu," random_state is :",max_rs)
```

```python
# MAKE FUNCTION FOR TRAIN & TEST MODEL

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score

from sklearn.model_selection import cross_val_score as CVS
from sklearn.model_selection import KFold
kfold= KFold(n_splits=5, shuffle=True, random_state=195)


    
def model_score(model,xtrain,ytrain,x_test,y_test):
    model.fit(xtrain,ytrain)
    pred= model.predict(xtrain)
    y_pred= model.predict(x_test)
    print('training_score =', accuracy_score(ytrain,pred))
            
    print('test_accuracy_score=', accuracy_score(y_test,y_pred))
    print('classification_report \n', classification_report(y_test,y_pred))
    print('confusion_matrix\n',confusion_matrix(y_test,y_pred))


def cross_val(model,xtrain,ytrain,algo):
        
    cv_score= CVS(model,xtrain,ytrain,cv= kfold,)
    print('cross_val_score with', algo,'=', cv_score.mean())
    
```

```python
# LOGISTIC REGRESSION

# we check above random_state =195 is best for splitting
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = 195,test_size=0.20,stratify=y)


lr=LogisticRegression()
 
model_score(lr,x_train,y_train,x_test,y_test)
```

```python
# IN TRAINING PROCESS
# finding best random_state for  'SVC'

from sklearn.svm import SVC

max_acu= 0
max_rs= 0
for r_state in range(50,300):
    x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = r_state,test_size=0.20,stratify=y)
    svm = SVC()
    svm.fit(x_train,y_train)
    y_pred=svm.predict(x_test)
    accu = accuracy_score(y_test,y_pred)
        #print("random state :",r_state," and f1 score: ",tmp)
        
    if accu>max_acu:
        max_acu=accu
        max_rs= r_state
        print("maximum accuracy :",max_acu," random_state is :",max_rs)

```

```python
# lets use SVM CLASSIFICATION

# we check above random_state = 170 is best for splitting
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = 170,test_size=0.20,stratify=y)


svc = SVC()

model_score(svc,x_train,y_train,x_test,y_test)
```

```python
# IN TRAINING PROCESS
          # finding best random_state for  'decisintree c_'
        
from sklearn.tree import DecisionTreeClassifier

max_acu= 0
max_rs= 0
for r_state in range(50,300):
    x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = r_state,test_size=0.20,stratify=y)
    dt = DecisionTreeClassifier()
    dt.fit(x_train,y_train)
    y_pred = dt.predict(x_test)
    accu = accuracy_score(y_test,y_pred)
        #print("random state :",r_state," and f1 score: ",tmp)
        
    if accu>max_acu:
        max_acu=accu
        max_rs= r_state
        print("maximum accuracy :",max_acu," random_state is :",max_rs)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

xtrain,xtest,ytrain,ytest=train_test_split(x, y,random_state = 170,test_size=0.20,stratify=y)


rf= RandomForestClassifier()

param= {'n_estimators':[180,300],'criterion':['gini','enetrop'],'max_depth':[2,3,4],'min_samples_leaf':[1,2,3,4]}

rand= RandomizedSearchCV(rf, param, cv=5, n_jobs=-1)
rand.fit(xtrain,ytrain)
rand.best_params_
```

```python
cls= rand.best_estimator_
cls.fit(xtrain,ytrain)
pred= cls.predict(xtest)

print(accuracy_score(ytest, pred))
```
