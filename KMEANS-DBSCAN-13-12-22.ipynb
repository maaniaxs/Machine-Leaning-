# MAKING A CLUSTRING MODEL_0
* on sklearn datasets make_moons 

```python
#
from sklearn.cluster import DBSCAN, KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score as ars
from sklearn.preprocessing import StandardScaler
```

```python
feature, label = make_moons(n_samples=250,noise=0.05, random_state=42)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(feature)
```

```python
kmeans = KMeans(n_clusters=2)

dbscan = DBSCAN(eps=0.3)
```

```python
kmeans.fit(scaled_data)

dbscan.fit(scaled_data)
```

```python
kmean_silhouette = silhouette_score(scaled_data, kmeans.labels_).round(2)
dbscan_silhouette = silhouette_score(scaled_data, dbscan.labels_).round(2)
print(kmean_silhouette)
print(dbscan_silhouette)
```

```python
kmean_ari = ars(label, kmeans.labels_)   # cheking adjusted_rand_score
dbscan_ari = ars(label, dbscan.labels_)
print('ARI kmeans', round(kmean_ari, 2))
print('ARI dbscan', round(dbscan_ari, 2))
```

```python
import pandas as pd
data = pd.DataFrame({'A':feature[:,0], 'B':feature[:,1], 'label':label, 'kmeans':kmeans.labels_, 'dbscan':dbscan.labels_})
data.head()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(x='A',y='B', data=data, hue='label')
plt.title('original data')
plt.subplot(1,3,2)
sns.scatterplot(x='A',y='B', data=data, hue='kmeans')
plt.title('kmeans predicted data')
plt.subplot(1,3,3)
sns.scatterplot(x='A',y='B', data=data, hue='dbscan')
plt.title('dbscan predicted data')
plt.show()
```

# Making clustering model_2
* at sklearn dataset make_blobs

```python
feature_2, label_2 = make_blobs(n_samples=250,cluster_std=1.2, random_state=42)
```

```python
kmeans_2 = KMeans(n_clusters=3)     # fit the KMeans model 

dbscan_2 = DBSCAN(eps=0.3)         # fit the dbscan model
```

```python
kmeans_2.fit(scaled_data_2)    # fit the KMeans model

dbscan_2.fit(scaled_data_2)   # fit the dbscan model
```

```python
kmean_silhouette_2 = silhouette_score(scaled_data_2, kmeans_2.labels_).round(2)
dbscan_silhouette_2 = silhouette_score(scaled_data_2, dbscan_2.labels_).round(2)
print(kmean_silhouette_2)
print(dbscan_silhouette_2)
```

```python
kmean_ari = ars(label_2, kmeans_2.labels_)   # cheking adjusted_rand_score
dbscan_ari = ars(label_2, dbscan_2.labels_)
print('ARI kmeans', round(kmean_ari, 2))
print('ARI dbscan', round(dbscan_ari, 2))
```

```python
import pandas as pd
data_2 = pd.DataFrame({'A':feature_2[:,0], 'B':feature_2[:,1], 'label':label_2, 'kmeans':kmeans_2.labels_, 'dbscan':dbscan_2.labels_})
data_2.head()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(x='A',y='B', data=data_2, hue='label')
plt.title('original data')
plt.subplot(1,3,2)
sns.scatterplot(x='A',y='B', data=data_2, hue='kmeans')
plt.title('kmeans predicted data')
plt.subplot(1,3,3)
sns.scatterplot(x='A',y='B', data=data_2, hue='dbscan')
plt.title('dbscan predicted data')
plt.show()
```
