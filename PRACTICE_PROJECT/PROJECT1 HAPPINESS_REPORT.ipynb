```python
x=df1.drop(['Happiness Score', 'Standard Error','Generosity'], axis=1)
y= df1['Happiness Score']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
```

```python
# STANDARDIZATION THE DATA

sc= StandardScaler()
scaler= sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y , test_size=0.25,random_state=119)
```

```python

def model_score(clf,xtrain,xtest,ytrain,ytest):
    clf.fit(xtrain,ytrain)
    pred_train= clf.predict(xtrain)
    pred_test= clf.predict(xtest)
    print("regression_score", clf.score(xtest,ytest),'\n')
    print('mean_squared_error= \n', MSE(ytest,pred_test),'\n')
          
    print('mean_absolute_error= \n', MAE(ytest,pred_test))
```

```python
# model fit with LinearRegression

lr= LinearRegression()
lr.fit(x_train,y_train)

lr_pred = lr.predict(x_test)
```

```python
model_score(lr,x_train, x_test, y_train, y_test)
```

```python
# MODEL FIT WITH  SVM
from sklearn.svm import SVR

svr = SVR()
```

```python
svr.fit(x_train, y_train)

svr_pred = svr.predict(x_test)

model_score(svr,x_train, x_test, y_train, y_test)
```

```python
gbr= GradientBoostingRegressor()
```

```python
gbr.fit(x_train,y_train)

gbr_pred= gbr.predict(x_test)

model_score(gbr,x_train, x_test, y_train, y_test)
```
