```python
# REMOVING  OUTLIER BY ZSCORE METHOD

from scipy.stats import zscore
z_scores=zscore(df[['Age','Fnlwgt']])
df_new = df.loc[(abs(z_scores)<2.9).all(axis=1)]

print('original dataset shape', df.shape)
print('new dataset shape', df_new.shape)
```

```python
#
x= df_new.drop('Income', axis=1)
```

```python
# make sure to multicollinearity is or not, use 'variance_inflation_factor' method to check  multicollinearity 

from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
features= x.columns
vif_value= [vif(x.values, i) for i in range(len(x.columns))]

pd.DataFrame({'features':features,'vif_value':vif_value})   #
# 
# multicollinearity problem in this dataset , some feature have to be removed
```

```python
x = x.drop([ 'Native_country', 'Education_num'], axis=1)   # input_feature

y= df_new['Income']  #   target_feature
```

```python


y.value_counts()    # imbalanced dataset
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.metrics import *
```

```python
#  SCALING THE DATASET 
sc= StandardScaler()
scaled= sc.fit_transform(x)

scaled= pd.DataFrame(scaled, columns=x.columns)
```

```python
# balanced the dataset with SMOTE 
x_train,x_test, y_train,y_test= train_test_split(scaled,y, test_size=0.30, random_state= 49)

from imblearn.over_sampling import SMOTE
smp= SMOTE(0.60)
xtrain, ytrain= smp.fit_resample(x_train,y_train)

print(y_train.value_counts(),)
print('\n after over_sampling \n', ytrain.value_counts())
```

```python
from sklearn.model_selection import cross_val_score as CVS
from sklearn.model_selection import KFold
kfold= KFold(n_splits=30, shuffle=True, random_state=49)
```

```python
# IN TRAINING PROCESS                 # finding best random_state for  MODEL

#max_acu= 0
#max_rs= 0
def random_state(model,xtrain,ytrain):
    max_acu= 0
    max_rs= 0
    for r_state in range(50,92):
        x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = r_state,test_size=0.30 )
    
        model.fit(x_train,y_train)
        y_pred=model.predict(x_test)
        accu = accuracy_score(y_test,y_pred)
        #print("random state :",r_state," and f1 score: ",tmp)
        
        if accu>max_acu:
            max_acu=accu
            max_rs= r_state
            print("maximum accuracy :",max_acu," random_state is :",max_rs)
```

```python
def model_score(model,x_train,y_train,x_test,y_test):
    model.fit(x_train,y_train)
    pred= model.predict(x_train)
    y_pred= model.predict(x_test)
    print('training_score =', accuracy_score(y_train,pred))
    print('test_accuracy_score=', accuracy_score(y_test,y_pred))
    
    print('classification_report \n', classification_report(y_test,y_pred))
    print('roc_auc_score =', roc_auc_score(y_test,y_pred))
    
def cross_val(model,x_train,y_train,algo):
    
    cv_score= CVS(model,x_train,y_train,cv= kfold,)
    print('cross_val_score with', algo,'=', cv_score.mean())
```

```python
#           RANDOMFOREST-CLASSIFIER

from sklearn.ensemble import *
RF = RandomForestClassifier()
random_state(RF ,xtrain,ytrain)
```

```python
#           RANDOMFOREST-CLASSIFIER
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 75,test_size=0.30 )
model_score(RF,x_train,y_train,x_test,y_test)
```

```python
#     SGD-Classifier 

from sklearn.linear_model import SGDClassifier
SGD= SGDClassifier()
random_state(SGD ,xtrain ,ytrain)
```

```python
#           SGD-Classifier
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 69,test_size=0.30 )
model_score(SGD,x_train,y_train,x_test,y_test)
```

```python
#                  let's use KNN-classifier

from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
random_state(KNN ,xtrain ,ytrain)
```

```python
#           knn-classifier
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 60,test_size=0.30 )
model_score(KNN,x_train,y_train,x_test,y_test)
```

```python
#  LET'S USE  EXTRATREES-CLASSIFIER 

TC = ExtraTreesClassifier(bootstrap=True,oob_score=True,)
random_state(TC ,xtrain,ytrain)
```

```python
#           EXTRATREES-CLASSIFIER
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 75,test_size=0.30 )
model_score(TC,x_train,y_train,x_test,y_test)
```

```python
#  CROSS-VALIDATION 
#       
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 75,test_size=0.30 )
cross_val(TC,x_train,y_train,'EXTRATREES-CLASSIFIER')                           ## CROSS_VALIDATION WITH EXTRATREES-CLASSIFIER
print('-'*15)

#        
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 75,test_size=0.30 )
cross_val(RF,x_train,y_train,'RANDOMFOREST-CLASSIFIER')                       ##  Cross_validation with  RandomForest-Classifier
print('-'*15)

#          
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 69,test_size=0.30 )
cross_val(SGD,x_train,y_train,'SGD-Classifier')                                ##  Cross_Validation with SGD-Classifier  
print('-'*15)

#
x_train,x_test,y_train,y_test=train_test_split(xtrain, ytrain,random_state = 60,test_size=0.30 )
cross_val(KNN,x_train,y_train,'KNN-Classifier')                                ##  Cross_Validation with SGD-Classifier
print('-'*15)

#
GB= GradientBoostingClassifier()
GB.fit(x_train, y_train)
cross_val(GB, x_train,y_train,'GradienBoosting-Classifier')                    ### cross_validation  with  GradientBoosting
```
